```python
nh3 = pd.read_excel('nh3bg.xlsx')
```

```python
nh3 = nh3.iloc[1: , :]
```

```python
nh3.rename(columns = {'Unnamed: 0':'Td', 'Unnamed: 1':'NH3'}, inplace = True)
```

```python
nh3.head(10)
```

```python
from sklearn.preprocessing import StandardScaler
X = nh3.drop('Td',axis=1)
y = nh3['Td']
scaler = StandardScaler()
X = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
nh3.head(10)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
rfc = RandomForestRegressor()
params = {'n_estimators':[100,200,300,400,500,600,700,800,900,1000]}
grid_model = GridSearchCV(rfc, params,verbose=2)
grid_model.fit(X_train,y_train)
pred = grid_model.predict(X_test)
print('Random Forest accuracy is --> ',r2_score(y_test,pred)*100)
```
